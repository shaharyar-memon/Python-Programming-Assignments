```python
# Question no. 1 
# Answer : Object Orented Programming is a way of programming in which we combine a group of related variables and ...
# ... and functions into a unit. This unit is called an 'Object'. The variables are referred to as 'Attributes/Properties'...
# ... and the functions as 'Methods. For example, a Car is an object having attributes such as color, model and company...
# ...  and having methods such as start(), stop() and others.

# Question no. 2
# Answer : Some benefits of OOP are 
# 1.The code is organized and easier to comprehend 
# 2.The code is re-useable (Inheritance)
# 3.The code is easier to maintain
# 4.The code is more secure by means of Encapsulation 

# Question no. 3
# Answer : A function is a piece of code that is called by name. It can be passed data to operate on and can...
# ... optionally return data. A method is a piece of code that is called by a name that is associated with...
# ... an object. A method is able to operate on data that is contained within the class.

# Question no. 4
# Answer : A 'Class' is a blueprint for creating an object. It contains varibles/attributes and behaviour in form of methods.
# an 'Object' is a software bundle of variables and related methods. 'Attributes' are data stored inside a class or...
# ... instance and represent the state or quality of the class or instance. 'Behaviour' or a class determines how an...
# ... instance of that class operates. In simple, variables inside a class are called 'Attributes' and functions inside...
# ... a class are called 'Methods' that define the behaviour of the object.

# Question no. 5
# Class
class car():
    def __init__(self, color, model, name, company, transmission):
        # Five attributes
        self.color = color
        self.name = name
        self.model = model
        self.company = company
        self.transmission = transmission
    
    # Three Methods
    def repaint_car(self, newColor):
        self.color = newColor
        print('\nThe car has been repainted with color:', self.color)
        
    def car_details(self):
        print('\n-- Car Details --\n', self.transmission, self.company, self.name, 'of year', self.model, 'and color', self.color)

    def change_car(self, newColor, newModel, newName, newCompany, newTransmission):
        self.color = newColor
        self.model = newModel
        self.name = newName
        self.company = newCompany
        self.transmission = newTransmission
        
# Five instances 
car1 = car('Red', '2006', 'RX8', 'Mazda', 'Automatic')
car2 = car('Black', '2019', 'Corolla', 'Toyota', 'Automatic')
car3 = car('Yellow', '2013', 'Mustang', 'Ford', 'Manual')
car4 = car('Grey', '2018', 'Civic', 'Honda', 'Automatic')
car5 = car('White', '2011', 'Cultus', 'Suzuki', 'manual')

print('', car1, '\n', car2, '\n', car3, '\n', car4, '\n', car5)
car1.car_details()
car2.car_details()
car2.repaint_car('Silver')
car2.car_details()

print('\n---------------------------------------------------------')
print('\nCar three - old')
car3.car_details()
car3.change_car('Blue', '2019', 'Aqua', 'Toyota', 'Automatic')
print('\nCar three - new')
car3.car_details()
```
